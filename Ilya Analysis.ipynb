```python
full_convo = pd.merge(convos, review, how = 'left', left_on= "id", right_on="conversation_id")
```

```python
# full_convo = full_convo[full_convo['cancelled_at']]
full_convo = full_convo[full_convo['booked_at'].notnull() & full_convo['cancelled_at'].isnull()]
```

```python
full_convo =full_convo.drop(columns = ['id_x', 'id_y','content'])
```

```python
full_convo
```

```python
full_convo_1 = full_convo[full_convo['reviewer_id'].notnull()]
```

```python
sum(full_convo_1['reviewer_id'] == full_convo_1['requester_id'])
```

```python
sum(full_convo_1['reviewer_id'] != full_convo_1['requester_id'])
```

```python
test1 = full_convo_1[full_convo_1['requester_id'] == full_convo_1['reviewer_id']][['stars', 'reviewer_id']]
test2 = full_convo_1[full_convo_1['requester_id'] != full_convo_1['reviewer_id']][['stars', 'reviewer_id']]

plt.hist(test1.iloc[:, 0], align = 'left', label = 'Owners')
plt.hist(test2.iloc[:,0], label = 'Providers')
plt.legend()
plt.xlabel("Stars")
plt.ylabel("Number of Reviews")
plt.title("Stars Frequency per Reviewer Type")
# # test1.iloc[:,0]

display(test1.iloc[:, 0].describe(), test2.iloc[:, 0].describe())

#Based on frequency of responses, owner is likely to provide review and leaves better reviews.
```
